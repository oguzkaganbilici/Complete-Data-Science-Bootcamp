# Multicollinearity

```python
data_cleaned.columns.values
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
variables = data_cleaned[["Mileage", "Year", "EngineV"]] # we input the columns that we check there is multicollinearity
vif = pd.DataFrame()
vif["VIF"] = [variance_inflation_factor(variables.values, i) for i in range(variables.shape[1])]
vif["Features"] = variables.columns
```

```python
vif
```

- #### VIF = 1: no multicollinearity
- #### we are trying to find 1< VIF <5 values.  but in some sources it says, 1< VIF <10 is ok.
- #### we can see that year may have some multicollinearity, so we will drop it out

```python
data_no_multicoll = data_cleaned.drop(["Year"], axis = 1)
data_no_multicoll
```
